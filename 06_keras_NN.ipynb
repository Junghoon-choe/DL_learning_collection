# import Module

```
import keras
import numpy as np
```

#1. 데이타 생성

```
x = np.array([0,1,2,3,4])
y = 2 * x + 1
print(x)
print(y)
```

#2. 케라스로 모델 생성

```
model = keras.models.Sequential()
# Dense()로 만들것임... *args (여러개의 변수가 들어갈수 있다는 말), **kwargs () **이 나왔을 경우 키워드를 예시를 통해 알아봐야한다.
model.add(keras.layers.Dense(1,input_shape=(1,))) # 출력노드가 앞에 나오고, 입력노드가 뒤에나옴. (채널을 말함).
```

units : 히든레이어 안의 노드 갯수를 말한다  
** kwargs() : 이런 변수는 없다. input_shape=(1,)

#3. compile() 을 사용해서 optimizer 정의

```
model.compile(optimizer=keras.optimizers.SGD(0.01), loss='MSE', metrics=['accuracy'])
```

#4. Model 학습진행

```
model.fit(x[:4], y[:4], epochs=100)# verbose=0 하면 출력 메세지가 안나옴.
print(x[:4])
print(y[:4])
```

#5. predict() 예측하기 

```
print(f'prediction : {model.predict(x[3:]).flatten()}')
```

#6. evaluation and accuracy

```
loss_function = model.evaluate(x[3:], y[3:])
print(f'한번에 출력 : {loss_function}')
print(f'나눠서 출력 : loss:{loss_function[0]}, accuracy:{loss_function[1]}')
```

evaluate()은 loss 값과 accuracy값 2개를 반환한다.
